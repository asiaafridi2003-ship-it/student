```python
TEMP_PLEASANT_MAX = 28
TEMP_PLEASANT_MIN = 20
HUMIDITY_PLEASANT_MAX = 60
WIND_PLEASANT_MAX = 15

temp = float(input("Enter temperature (in Celsius): "))
humidity = float(input("Enter humidity (in %): "))
wind = float(input("Enter wind speed (in kph): "))

if (temp >= TEMP_PLEASANT_MIN and temp <= TEMP_PLEASANT_MAX) and \
   (humidity < HUMIDITY_PLEASANT_MAX) and \
   (wind < WIND_PLEASANT_MAX):

    classification = "Pleasant"

elif (temp > 35 or temp < 5) or \
     (humidity > 85) or \
     (wind > 35):

    classification = "Harsh"

else:
    classification = "Normal"

print("\n--- Weather Classification ---")
print(f"Temperature: {temp}°C, Humidity: {humidity}%, Wind: {wind} kph")
print(f"Classification: {classification}")
```

```python
count = 0
total = 0

while True:
    num = int(input("Enter a number: "))

    if num <= 0:
        break # Exit the loop when a negative number is entered

    total += num
    count += 1

# Check if any non-negative numbers were entered before calculating the average
if count > 0:
    average = total / count
    print(f"The average of the {count} numbers is: {average:.2f}")
else:
    print("No non-negative numbers were entered.")
```

```python
def is_palindrome(n):
    original_n = n
    reversed_n = 0
    if n < 0:
        return False
    while n > 0:
        digit = n % 10
        reversed_n = (reversed_n * 10) + digit
        n = n // 10
    return original_n == reversed_n
print(f"Is 121 a palindrome? {is_palindrome(121)}")
print(f"Is 1234 a palindrome? {is_palindrome(1234)}")
print(f"Is 5005 a palindrome? {is_palindrome(5005)}")
```
