```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def display_stats(self):
        area = self.length * self.width
        perimeter = 2 * (self.length + self.width)
        print(f"Area: {area}, Perimeter: {perimeter}")

rect = Rectangle(10, 5)
rect.display_stats()
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks  # Expecting a list of 3 numbers

    def print_average(self):
        avg = sum(self.marks) / len(self.marks)
        print(f"{self.name}'s Average: {avg}")

s1 = Student("Alice", [85, 90, 78])
s1.print_average()
```

```python
class StudentResult:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def check_status(self):
        status = "Passed" if self.marks >= 50 else "Failed"
        print(f"{self.name} has {status}")

s2 = StudentResult("Bob", 45)
s2.check_status()
```

```python
class Account:
    def __init__(self, acc_no, balance):
        self.acc_no = acc_no
        self.balance = balance

    def debit(self, amount):
        self.balance -= amount
        print(f"Debited: {amount}")

    def credit(self, amount):
        self.balance += amount
        print(f"Credited: {amount}")

    def print_balance(self):
        print(f"Account {self.acc_no} Balance: {self.balance}")

acc = Account(101, 5000)
acc.credit(1000)
acc.debit(500)
acc.print_balance()
```

```python
class Employee:
    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def display_annual_salary(self):
        annual = self.salary * 12
        print(f"Employee: {self.name}, Annual Salary: {annual}")

emp = Employee(1, "John Doe", 3000)
emp.display_annual_salary()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. Remaining balance: ${self.__balance}")
        elif amount > self.__balance:
            print("Insufficient funds!")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        # Controlled way to view the balance
        return f"Account owner: {self.owner} | Balance: ${self.__balance}"

# --- Testing the System ---
account = BankAccount("Alex", 1000)

account.deposit(500)   # Works
account.withdraw(200)  # Works
print(account.get_balance())

# This will trigger an error because __balance is private
# print(account.__balance)
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient funds")

account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
```

```python

class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def display(self):
        print(f"Car: {self.brand} {self.model}")

my_car = Car("Toyota", "Corolla")
my_car.display()
```
